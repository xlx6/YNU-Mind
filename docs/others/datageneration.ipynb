```python
import json

# 加载爬取的数据
with open("teacher_details_cleaned.json", "r", encoding="utf-8") as f:
    teacher_data = json.load(f)

# 初始化问答对列表
qa_pairs = []

# 遍历每位老师的信息
for teacher in teacher_data:
    name = teacher.get("名字", "未知老师")
    position = teacher.get("职位")
    contact = teacher.get("联系方式")
    address = teacher.get("地址")
    bio = teacher.get("个人简介")
    research_direction = teacher.get("研究方向", bio)  # 研究方向可能隐含在简介中
    papers = teacher.get("期刊论文", "暂无相关信息")
    projects = teacher.get("科研项目", "暂无相关信息")
    patents = teacher.get("主要专利", "暂无相关信息")

    # 基础信息问题
    if contact:
        qa_pairs.append({
            "instruction": f"{name}老师的联系方式是什么？",
            "input": "",
            "output": contact
        })
    if position:
        qa_pairs.append({
            "instruction": f"{name}老师的职位是什么？",
            "input": "",
            "output": position
        })
    if address:
        qa_pairs.append({
            "instruction": f"{name}老师的办公地址是什么？",
            "input": "",
            "output": address
        })

    # 研究方向和个人简介
    if research_direction:
        qa_pairs.append({
            "instruction": f"{name}老师的研究方向是什么？",
            "input": "",
            "output": research_direction
        })
    if bio:
        qa_pairs.append({
            "instruction": f"介绍一下{name}老师。",
            "input": "",
            "output": bio
        })

    # 论文相关问题
    if papers != "暂无相关信息":
        qa_pairs.append({
            "instruction": f"{name}老师发表过哪些论文？",
            "input": "",
            "output": papers
        })

    # 项目相关问题
    if projects != "暂无相关信息":
        qa_pairs.append({
            "instruction": f"{name}老师主持过哪些科研项目？",
            "input": "",
            "output": projects
        })

    # 专利相关问题
    if patents != "暂无相关信息":
        qa_pairs.append({
            "instruction": f"{name}老师有哪些专利？",
            "input": "",
            "output": patents
        })

    # 关系型问题
    if research_direction:
        qa_pairs.append({
            "instruction": f"与{research_direction}最相关的老师是谁？",
            "input": "",
            "output": f"{name}老师"
        })

# 保存生成的数据集
with open("teacher_qa_dataset.json", "w", encoding="utf-8") as f:
    json.dump(qa_pairs, f, ensure_ascii=False, indent=4)

print(f"问答数据集生成完成，共包含 {len(qa_pairs)} 个问答对！")
```
